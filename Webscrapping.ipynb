## IMDB Web Scraping

#### Importing all the libraries

```python
import pandas as pd 
import numpy as np
import requests
from bs4 import BeautifulSoup
```

#### URL of the site

```python
url = 'https://www.imdb.com/list/ls000021718/'
```

#### Use headers that stimulates request coming from web-browser

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
response = requests.get(url, headers=headers)
```

```python
with requests.Session() as session:
    response = session.get(url, headers=headers)
```

```python
response
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
movie_name = []
year = []
time = []
genre = []
rating = []
votes = []
gross = []
```

```python
movie_data = soup.findAll('div', attrs = {'class': 'lister-item mode-detail'})
```

```python
for store in movie_data:
    all_movies_title = store.h3.a.text
    movie_name.append(all_movies_title)
    
    release_year = store.find('span', class_="lister-item-year text-muted unbold").text.replace('(','').replace(')','')
    year.append(release_year)
    
    runtime = store.find('span', class_="runtime").text.replace(' min','')
    time.append(runtime)
    
    genres = store.find('span', class_="genre").text.replace('[\n','').replace(' ]','').replace('\n','')
    genre.append(genres)
    
    ratings = store.find('span', class_="ipl-rating-star__rating").text
    rating.append(ratings)
    
    value = store.find_all('span', attrs = {'name': 'nv'})
    
    vote = value[0].text
    votes.append(vote)
    
    grosses = value[1].text if len(value) >1 else '*****'
    gross.append(grosses)
```

```python
movies_data = pd.DataFrame({'Title':movie_name, 'Release_Year':year, 'Duration(min)':time, 'Genre':genre, 'Ratings':rating, 'Votes':votes, 'Gross':gross})
```
